```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder as LE
from sklearn.decomposition import PCA
import xgboost as xgb
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.shape
```

```python
test.shape
```

```python
zero_var_train = train.var()[train.var()==0] 
zero_var_train
```

```python
train.drop(zero_var_train.index.values, axis=1, inplace=True)  
```

```python
zero_var_test = test.var()[test.var()==0]
zero_var_test
```

```python
test.drop(zero_var_test.index.values, axis=1, inplace=True)  
```

```python
train.isna().sum().sum()
```

```python
test.isna().sum().sum()
```

```python
for i in train.columns:
    print('\n' + train[i].name)
    print(train[i].unique())
    print("Total no of unique values are", train[i].nunique())
```

```python
for i in test.columns:
    print('\n' + test[i].name)
    print(test[i].unique())
    print("Total no of unique values are", test[i].nunique())
```

```python
le = LE()

train_object = train.select_dtypes(include='object').columns
for i in train_object:
    train[i] = le.fit_transform(train[i])
    
test_object = test.select_dtypes(include='object').columns
for i in test_object:
    test[i] = le.fit_transform(train[i])
```

```python
X_train = train.drop('y', axis=1)
y_train = train.y

pca = PCA(n_components=2)
train_pca = pca.fit_transform(X_train)
test_pca = pca.fit_transform(test)
```

```python
model = xgb.XGBRegressor()
model.fit(train_pca, y_train)
```

```python
model.predict(test_pca)
```
